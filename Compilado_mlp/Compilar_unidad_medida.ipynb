```python
import pandas as pd
import openpyxl
import os



directory_path = r'G:\.shortcut-targets-by-id\1ugEph6zRyoKHbwpiYuXbB3Pj0UXueIKJ\09. Contrato CS115 EIA 2024\03.WIP\03.Hidroquímica\2. Actualizacion BDD\2. Nuevo - Mauro\1. Certificados\1. Revision final'


# Crear un DataFrame vacío para almacenar los datos de los archivos Excel
data_parametros = pd.DataFrame()
data_compared = pd.DataFrame()

data_compared_1 = pd.DataFrame()

excel_files = [file for file in os.listdir(directory_path) if file.endswith(('.xlsx'))]

# Procesar cada archivo Excel encontrado
for file in excel_files:
    file_path = os.path.join(directory_path, file)  # Construir la ruta completa al archivo
    xls = pd.ExcelFile(file_path)
    wb = openpyxl.load_workbook(file_path)

    for hoja_archivo in xls.sheet_names:
        hoja = wb[hoja_archivo]
        rango = hoja['C133:AZ133']

        ultima_celda_con_datos = None
        for celda in rango[0]:
            if celda.value is not None:
                ultima_celda_con_datos = celda

        if ultima_celda_con_datos:
            columna = ''.join(filter(str.isalpha, ultima_celda_con_datos.coordinate))
            print(f'La última columna con datos es >>> {columna} <<< , en la hoja: >>> {hoja_archivo} <<<')
            
    
            data_df = pd.read_excel(xls, sheet_name= hoja_archivo, skiprows=141, nrows=120, usecols=f"B:{columna}", header= None)  
            data_df.reset_index(drop=False)
            data_compared_1 = pd.DataFrame()
            data_compared_1 = data_df[[1, 2]].copy()

            data_compared_1['Ruta'] = file_path
            data_compared_1['Hoja'] = hoja_archivo
            data_compared_1['Archivo'] = file
            data_compared_1.columns = ['Parametros', 'Unidad_medida', 'Ruta', 'Hoja', 'Archivo']
            
            data_df.drop(data_df.columns[1], axis=1, inplace=True)
            
           
            
            # data_pivoted = data_df.T
            

            # # PARAMETROS COMO COLUMNAS
            # new_header = data_pivoted.iloc[0]  
            # data_pivoted = data_pivoted[1:]  
            # data_pivoted.columns = new_header

            # # data_parametros.reset_index(drop=True, inplace=True)
            # # data_pivoted.reset_index(drop=True, inplace=True)
            # data_pivoted.reset_index(drop=True, inplace=True)

            # # data_identificacion total
            # data_parametros = pd.concat([data_parametros, data_pivoted], ignore_index=True)
            # # data_parametros = pd.concat([data_parametros, data_df], ignore_index=True)
            # for column in data_parametros.columns:
            #     data_parametros[column] = pd.to_numeric(data_parametros[column], errors='coerce')
            #data_parametros = data_parametros.apply(pd.to_numeric, errors='coerce')
            data_compared = pd.concat([data_compared, data_compared_1], ignore_index=True)
        else:
            print(f'No hay datos en el rango C133:AZ133, para la hoja: {hoja_archivo}')
            
         
# data_parametros
data_compared
```
