```python
import re
#cambiar columna Smp a float
# df['Smp'] = df['Smp'].astype(float)
# #CAMBIAR PUNTOS COMAS X PUNTOS  
# df['Smp'] = df['Smp'].str.replace(',', '.')
# df['Smp'] = df['Smp'].apply(lambda x: re.sub(r'(?<!^)\.', '', str(x), count=1))

# df['Smp'] = df['Smp'].apply(lambda x: str(x).replace(',', '.'))
# df['Smp'] = df['Smp'].astype(float)

# # Aplicar la condición y transformación
# df['Smp'] = df['Smp'].apply(lambda x: x / 1000 if x > 1000 else x)

def remove_additional_points(value):
    # Convertimos el valor a string
    value_str = str(value)
    # Si hay más de un punto en la cadena
    if value_str.count('.') > 1:
        # Partimos la cadena en el primer punto y unimos el resto removiendo los puntos adicionales
        parts = value_str.split('.', 1)
        return parts[0] + '.' + parts[1].replace('.', '')
    else:
        # Si hay cero o un punto, devolvemos el valor original
        return value_str

# Aplicar la función a la columna 'Smp'
df['Smp'] = df['Smp'].apply(remove_additional_points)
```

```python
df['Smp'] = df['Smp'].apply(lambda x: str(x).replace(',', '.'))

df['Smp'] = df['Smp'].astype(float)

df['Smp'] = df['Smp'].apply(lambda x: x / 1000 if x > 45 else x)
df['Smp'] = df['Smp'].apply(lambda x: x / 1000 if -10000 <= x <= -15 else x)

df = df.drop_duplicates(subset=['Unnamed: 0'])
```

```python
import pandas as pd

# Supongamos que ya tienes el DataFrame 'df' cargado desde un archivo CSV o similar
# df = pd.read_csv('ruta/a/tu/archivo.csv')

# Verificar si 'Unnamed: 0' es el índice
if df.index.name != 'Unnamed: 0':
    # Asegurarnos de que 'Unnamed: 0' sea de tipo datetime
    df['Unnamed: 0'] = pd.to_datetime(df['Unnamed: 0'])
    #COLUMNA Smp como float
    
    
    # Establecer 'Unnamed: 0' como índice
    df.set_index('Unnamed: 0', inplace=True)

# Renombrar la columna de temperatura para mayor claridad
df.rename(columns={'Smp': 'temperature'}, inplace=True)

# Crear un DataFrame resampleado a intervalos de 1 día para calcular Tmax y Tmin puntuales
daily_max = df.resample('D').max()
daily_min = df.resample('D').min()

# Función para calcular el promedio de temperatura en un intervalo específico
def calculate_interval_avg(df, time_str, interval='45T'):
    start_time = pd.to_datetime(time_str)
    end_time = start_time + pd.Timedelta(interval)
    interval_df = df[start_time:end_time]
    return interval_df['temperature'].mean()

# Crear un DataFrame para almacenar resultados
results = []

# Iterar sobre cada día y calcular Tmax y Tmin puntuales y sus promedios
for date, max_temp in daily_max['temperature'].items():
    min_temp = daily_min.loc[date, 'temperature']
    
    # Encontrar las horas correspondientes a Tmax y Tmin puntuales
    max_temp_time = df[df['temperature'] == max_temp].index[0]
    min_temp_time = df[df['temperature'] == min_temp].index[0]
    
    # promedios en los intervalos de 45 minutos
    max_temp_avg = calculate_interval_avg(df, max_temp_time)
    min_temp_avg = calculate_interval_avg(df, min_temp_time)
    
    # Almacenar los resultados
    results.append({
        'date': date,
        'Tmax_puntual': max_temp,
        'Tmax_prom': max_temp_avg,
        'Tmin_puntual': min_temp,
        'Tmin_prom': min_temp_avg
    })

# Convertir los resultados en un DataFrame
results_df = pd.DataFrame(results)

# Mostrar los resultados
results_df
```
